# Form Finding for a Given Slope Study

The purpose of this notebook is to explore the relationship of the origin location and the geometric slope between endpoints. The end goal, is to come up with an algorithm that can find the origin location based on a desired slope and a maximum allowable force in the form. 


To look at the relationship, X, Y, and Z are defined by a given span. Then, the height of the midpoint, Y, is varied and the optimal origin is found through the form finding algorithm developed for geometric constraints. This origin for each Y height is plotted and there appears to be some sort of asymptotic behavior. 

```python
loads = [[0, -2], [3, -2], [0, -2]]
span = 8

X_loc = (np.cumsum(np.ones((1, len(loads)+2)))-1) *span/(len(loads)+1)

X = (0, 0)
Y = (span/2, 3)
Z = (span, 0)

poly = funicular_polygon()
poly.set_loads(loads)
poly.find_form(X_loc, X, Y, Z)
poly.get_geometry(X_loc)

fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].set_title('Force Diagram')
ax[1].set_title('Form Diagram')
poly.plot_force(ax=ax[0])
poly.plot_form(ax=ax[1])
for a in ax:
    a.axis('equal')
```

The figure above illustrates the load line that will be studied with a sample funicular form 

```python
y_range = [.5, 1, 2, 2.5, 3, 4, 5]

fig, ax = plt.subplots(1, len(y_range), figsize=(len(y_range)*3, 5))

for i, y in enumerate(y_range): 
    Y = (span/2, y)
    poly.find_form(X_loc, X, Y, Z)
    poly.plot_form(ax=ax[i])
    ax[i].set_title('Y Height = '+str(y))
    ax[i].scatter(Y[0], y, c='k')
    ax[i].axis('equal')
    
plt.suptitle('Iterating through forms with various Y heights')
```

```python
'''ITERATE THROUGH GEOMETRIES TO FIND ORIGINS'''

O = []
for y in range(-2000, 2000):
    if y !=0:
        Y = (span/2, y)
        o = poly.find_form(X_loc, X, Y, Z)
        
        O.append(o)
```

```python
'''APPROXIMATE'''
Fx = poly.load_line[-1, 0]
Fy = poly.load_line[-1, 1]
```

```python
origins = np.array(O)
xwin = 5
ywin = 10

o_plot = plt.scatter(origins[:, 0], origins[:, 1], s=5)
midy_plot = plt.plot([Fx/2-xwin, Fx/2+xwin], [Fy/2, Fy/2], 'k')
midx_plot = plt.plot([Fx/2, Fx/2], [Fy/2-ywin, Fy/2+ywin], 'g')


xp = np.arange(Fx/2-xwin, Fx/2+xwin, .1)
xp[xp==1]=0
appx_plot = plt.scatter(xp, -Fx/(xp-Fx/2)+Fy/2, s=1)
plt.legend([o_plot, appx_plot, midx_plot[0], midy_plot[0]], ['Origin Locations', 'Approx. Locations', 'X = '+str(Fx/2), 'Y = '+str(Fy/2)])



plt.xlim([Fx/2-xwin, Fx/2+xwin])
plt.ylim([Fy/2-ywin, Fy/2+ywin])
```

The figure above shows the asymptotes plotted along with the center of the load line. There appears to be some sort of asymptotic relationship following the form: 

$$ y=\frac{-F_{x}}{(x-\frac{F_{x}}{2})} + \frac{F_{y}}{2} $$

where $x$ is the x coordinate of the origin, $y$ is the y coordinate of the origin, $F_{x}$ is the resultant x-component of the resultant force, and $F_{y}$ is the y-component of the resultant force. 

While this equation holds for the current set up, it does not hold well when the load line is altered such that the inclined load is not the center. It also does not hold when there are other inclined loads, or the Z endpoint is changed, indicating that this formulation is incomplete.
